```python
import pandas as pd

# Ruta al archivo
file_path = r"C:\Users\Rodrigo\Desktop\Roxana Files\base_maestra\proyectos_consolidados_actualizados.xlsx"

# Cargar la base maestra actualizada
base_maestra = pd.read_excel(file_path, sheet_name='Base Maestra Actualizada')
```

```python
# Resumen general de la base de datos
eda_summary = base_maestra.describe(include='all')

# Guardar el resumen como un archivo
eda_summary.to_excel(r"C:\Users\Rodrigo\Desktop\Roxana Files\base_maestra\eda_summary.xlsx")
print("Resumen guardado en: eda_summary.xlsx")
```

```python
import matplotlib.pyplot as plt

# Distribucin de "Tiene fondecyt"
fondecyt_distribution = base_maestra['Tiene fondecyt'].value_counts(normalize=True)

plt.figure(figsize=(8, 6))
fondecyt_distribution.plot(kind='bar', title='Distribucin de "Tiene fondecyt"')
plt.xlabel('Tiene fondecyt')
plt.ylabel('Proporcin')
plt.show()

# Distribucin del nmero de proyectos
numero_proyectos_dist = base_maestra['numero_proyectos'].value_counts()

plt.figure(figsize=(8, 6))
numero_proyectos_dist.plot(kind='bar', title='Distribucin del nmero de proyectos por persona')
plt.xlabel('Nmero de Proyectos')
plt.ylabel('Frecuencia')
plt.show()
```
